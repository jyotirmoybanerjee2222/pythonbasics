```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
#normal distribution/gausian distribution
#measority of datapoints is clustered around mean

mu ,sigma = 0,0.1 

s=np.random.normal(mu,sigma,size=1000)
print("count of dps",len(s))
```

```python
plt.hist(s,30,density=True)
plt.show()
```

```python
sn.distplot(s)
```

```python
#log normal distribution >> whose logarithm is normal distributed
l=np.random.lognormal(mean=0,sigma=0.5,size=1000)
l=l[np.where(abs(l)<5)]
plt.hist(l,bins=100)
plt.show()
```

```python
# converting log normal distribution to normal distribution >> logtransformation
sn.distplot(np.log(l))
```

```python
#square root transformation
sn.distplot(np.sqrt(l))
```

```python
#chisquare distr ibution >> chisquare test
c=np.random.chisquare(df=1,size=10000)
c=c[np.where(abs(c)<10)]
plt.hist(c,bins=1000)

plt.show()
```

```python
# t distribution
#tail is thicker/larger
t=np.random.standard_t(df=1,size=10000)
t=t[np.where(abs(t)<10)]
plt.hist(t,bins=1000)
plt.show()
```

```python
# t distribution
#tail is thicker/larger
#degree of freedom =10
t=np.random.standard_t(df=10,size=10000)
t=t[np.where(abs(t)<10)]
plt.hist(t,bins=1000)
plt.show()
```

```python
# t distribution
#tail is  more thiner/larger
#dof=31
#approx to normal distribution
t=np.random.standard_t(df=31,size=10000)
t=t[np.where(abs(t)<10)]
plt.hist(t,bins=1000)
plt.show()
```

```python
#f distribution-> positive distribution
#dfnum is df for numerator,dfden=df for den
f= np.random.f(dfnum=1,dfden=1,size=10000)
f=f[np.where(abs(f)<10)]
plt.hist(f,bins=1000)
plt.show()
```
