```python
plt.plot(x,np.sin(x-0),color='blue')
plt.ylim(-1.5,1.5)
plt.xlim(1,15)
plt.grid()
plt.axis('equal')

plt.title("sin plot")
plt.show()
```

```python
plt.plot(x,np.sin(x-0),color='blue',label='sinx')
plt.plot(x,np.cos(x-0),color='green',label='cosx')
plt.ylim(-1.5,1.5)
plt.xlim(1,12)
plt.grid()

plt.xlabel("x-axIS")
plt.ylabel("Y-axIS")

plt.title("sin plot")
plt.legend()
plt.show()
```
